Q.2) Explain the purpose of constraints and how they help maintain data integrity in a database. Provide examples of common types of constraints?

**Solution : **

**Purpose of Constraints:**
* Maintain Data Integrity: Ensure data is accurate, consistent, and reliable by enforcing rules.
*  Prevent Errors: Limit invalid data input or operations.
* Enforce Business Rules: Ensure logical consistency and correctness.

**common types of constraints : **

•	Primary Key: Ensures each record is uniquely identified.

Example: emp_id INT PRIMARY KEY

•	Foreign Key: Ensures referential integrity between tables.

Example: FOREIGN KEY (dept_id) REFERENCES  departments(dept_id)

•	Unique: Ensures values in a column are unique.

 Example: email  VARCHAR(255)  UNIQUE

•	Check: Ensures values meet specific conditions.

 Example: age DECIMAL  CHECK(age > = 18)   

•	Not Null: Ensures a column cannot have NULL values.

Example: emp_name TEXT NOT NULL

•	Default: Provides a default value when none is provided.

Example: salary DECIMAL DEFAULT 30000

5)Explain the consequences of attempting to insert, update, or delete data in a way that violates constraints. Provide an example of an error message that might occur when violating a constraint?

**Solution : **

*Consequences of Violating Constraints:*

    *  Insert Violation: Data is rejected if it violates a constraint

    (e.g., inserting duplicate data in a UNIQUE column or NULL in a NOT NULL column).

    *  Update Violation: Updates fail if they violate constraints
    
    (e.g., changing a value to something invalid like a negative age when a CHECK constraint is applied).

    *  Delete Violation: Deletes are prevented if they would break referential integrity
    
    (e.g., deleting a row referenced by a foreign key).

### Normalisation & CTE

1. First Normal Form (1NF):

Question: Identify a table in the Sakila database that violates 1NF. Explain how you would normalize it to achieve 1NF.

Answer: A table violates 1NF if it contains repeating groups or columns with multiple values (e.g., multiple phone numbers stored in a single column).

    To achieve 1NF, ensure that:

        * All columns have atomic values (no multiple values in one column).

        * Remove any repeating groups and ensure each column contains a single value.

        * Create new tables for multi-valued attributes (like phone numbers in a separate table).


---


2. Second Normal Form (2NF):

Question: Choose a table in Sakila and describe how you would determine whether it is in 2NF. If it violates 2NF, explain the steps to normalize it.

Answer:

    * A table violates 2NF if it is in 1NF but contains partial dependencies (where a non-prime attribute depends only on a part of the composite primary key).

    * To check for 2NF, ensure that all non-prime attributes depend on the entire primary key.

    To normalize it:

        * Identify the partial dependencies.

        * Remove partial dependencies by splitting the table into separate tables.

---

3. Third Normal Form (3NF):

Question: Identify a table in Sakila that violates 3NF. Describe the transitive dependencies present and outline the steps to normalize the table to 3NF.

Answer: A table violates 3NF if it is in 2NF but contains transitive dependencies (where non-prime attributes depend on other non-prime attributes).

    To normalize it to 3NF:

        * Identify the transitive dependencies.

        * Remove transitive dependencies by creating new tables for those dependencies.

12. CTE for Recursive Search:

Question: Implement a recursive CTE to find all employees in the staff table who report to a specific manager, considering the reports_to column.

Answer:

WITH RECURSIVE EmployeeHierarchy AS (

  SELECT employee_id, first_name, last_name, reports_to

  FROM staff

  WHERE reports_to IS NULL  -- Start with top-level managers

  UNION ALL

  SELECT s.employee_id, s.first_name, s.last_name, s.reports_to

  FROM staff s

  INNER JOIN EmployeeHierarchy eh ON s.reports_to = eh.employee_id
)

SELECT * FROM EmployeeHierarchy;
