1.	What is a parameter?

ANS : A parameter is a variable that is part of the model's configuration and is learned from the data during training.

---


2.	What is correlation? What does negative correlation mean?

ANS : Correlation measures the linear relationship between two variables. Negative correlation means as one variable increases, the other decreases.

---

3.	Define Machine Learning. What are the main components in Machine Learning?

ANS : Machine Learning is a subset of AI that enables systems to learn from data and make decisions with minimal human intervention. Key components include data, models, algorithms, and evaluation metrics.

---


4.	How does loss value help in determining whether the model is good or not?

ANS : The loss value measures how well the model's predictions match the actual data. Lower loss values indicate a model that better fits the data.

5.	What are continuous and categorical variables?

ANS : Training a dataset involves using part of the data to teach a model, while testing involves using a separate part to evaluate its performance.

9.	What is a Test set?

ANS : A test set is a portion of the data set aside to assess the performance of a trained model, simulating how it would perform under real-world conditions.

---

10.	How do we split data for model fitting (training and testing) in Python?

15.	What is causation? Explain the difference between correlation and causation with an example.

ANS : Causation implies one event causes another. For example, smoking causes lung cancer. Correlation between two variables doesn't imply one causes the other, like ice cream sales and rate of drownings, which both increase in summer but don't cause each other.

16.	What is an Optimizer? What are different types of optimizers? Explain each with an example.

ANS : Optimizers adjust the attributes of the neural network like weights to minimize the loss function. Examples include SGD (updates parameters using the gradient of the loss function), Adam (adapts learning rates for each parameter), and RMSprop (designed to solve Adagrad's radically diminishing learning rates).
