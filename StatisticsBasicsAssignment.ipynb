Q.1)Explain the different types of data (qualitative
and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales?

**Answer:** Data can be categorized into qualitative (categorical) and quantitative (numerical) types. Qualitative data represent categories or labels and are subdivided into nominal and ordinal scales. For example, nominal data could include car colors (red, blue, green), and ordinal data could include rankings (first, second, third). Quantitative data represent quantities and are subdivided into interval and ratio scales. Interval data, like temperature in Celsius, have meaningful differences between values but no true zero, while ratio data, like weight in kilograms, have meaningful zeros, allowing for the comparison of magnitudes.

Q.2)What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

**Answer:** Measures of central tendency include the mean, median, and mode, which describe the center of a data set. The mean, or average, is used when data are normally distributed and there are no outliers (e.g., average temperature). The median is the middle value in an ordered list and is used when data are skewed or have outliers (e.g., median income). The mode is the most frequent value, useful in categorical data (e.g., most common birth month).

Q.3)Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

**Answer:** Dispersion refers to the spread of values around the central tendency. Variance measures the average squared deviations from the mean, providing a sense of how data are spread out. Standard deviation is the square root of variance and is in the same units as the data, making it easier to interpret. Both metrics help understand data variability; high values indicate wide dispersion, while low values suggest tight clustering.

Q.4)What is a box plot, and what can it tell you about the distribution of data?

**Answer:** A box plot is a graphical representation of data distribution summarized through five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It highlights the median, spread, skewness, and outliers. The interquartile range (IQR), calculated as Q3 - Q1, shows where the middle 50% of data lie. Outliers are often shown as points outside the 'whiskers'.

Q.5)Discuss the role of random sampling in making inferences about populations?

**Answer:** Random sampling is a method where each member of a population has an equal chance of being included in the sample. It minimizes bias and ensures that inferences and conclusions drawn from the sample are representative of the entire population. This is crucial in statistical hypothesis testing and confidence interval estimation.

Q.6)Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

**Answer:** Skewness refers to the asymmetry in the distribution of data. Positive skew (right-skewed) means tail on the right side is longer, negative skew (left-skewed) has a longer left tail. Skewness affects the mean's reliability as a measure of central tendency; in skewed distributions, the median is often a better measure of central tendency than the mean.


---


Q.7)What is the interquartile range (IQR), and how is it used to detect outliers?

**Answer:** The IQR is the difference between the third quartile (Q3) and the first quartile (Q1) in a data set, representing the range within which the middle 50% of scores fall. Outliers can be detected using the IQR; typically, any data point more than 1.5 IQRs below Q1 or above Q3 is considered an outlier.

Q.8)Discuss the conditions under which the binomial distribution is used.

**Answer:** The binomial distribution is used when the process meets the following conditions: there are a fixed number of trials, each trial has only two possible outcomes (success or failure), the probability of success is constant, and the trials are independent of each other. This distribution is useful for modeling scenarios like flipping a coin or quality control tests.

Q.9)Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule)?

**Answer:** The normal distribution is symmetrical and bell-shaped, centered around its mean. The empirical rule states that for a normal distribution, approximately 68% of the data falls within one standard deviation of the mean, 95% within two, and 99.7% within three standard deviations. This rule helps in assessing probabilities and understanding data spread.

Q.11)Explain what a random variable is and differentiate between discrete and continuous random variables?

**Answer:** A random variable is a variable whose value depends on the outcomes of a random phenomenon. A discrete random variable has a countable number of values (e.g., number of cars in a lot), while a continuous random variable can take any value within a given range (e.g., weight of a person).

Q.12)Provide an example dataset, calculate both covariance and correlation, and interpret the results?

**Answer:** Consider a dataset where two variables, X and Y, represent the hours studied and test scores of students. Covariance indicates the direction of the linear relationship between variables, while correlation measures both the strength and direction. If we calculate a positive covariance and a correlation close to 1, it suggests a strong positive relationship between hours studied and test scores.
