# Theory Questions

1. What is a random variable in probability theory?

Ans: A random variable is a variable whose value is determined by the outcome of a random phenomenon. It typically assigns a numerical value to each possible outcome of an experiment.


---


2. What are the types of random variables?

Ans: There are two main types of random variables: discrete and continuous. Discrete random variables have countable outcomes, while continuous random variables have outcomes that form a continuum.


---


3. What is the difference between discrete and continuous distributions?

Ans: Discrete distributions represent scenarios where outcomes are countable and distinct, while continuous distributions represent scenarios where outcomes can take any value within a continuous range.

---

4. What are probability distribution functions (PDF)?

Ans: A probability distribution function (PDF) describes the likelihood of different outcomes for a continuous random variable. It is a function whose integral over a particular interval gives the probability of the random variable falling within that interval.

---

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

Ans: The cumulative distribution function (CDF) of a random variable gives the probability that the variable is less than or equal to a certain value, whereas the probability distribution function (PDF) describes the probability density of a continuous random variable at a specific value.

---

6. What is a discrete uniform distribution?

Ans: A discrete uniform distribution is a probability distribution where all outcomes are equally likely. Each outcome has the same probability of occurring.

---

7. What are the key properties of a Bernoulli distribution?

Ans: A Bernoulli distribution is a discrete distribution having two possible outcomes, 1 ("success") and 0 ("failure"), with a single trial. It is characterized by the probability of success pp and the probability of failure 1−p1−p.

---

8. What is the binomial distribution, and how is it used in probability?

Ans: The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials, with each trial having the same probability of success. It is used to model scenarios where the outcomes are binary (success/failure) and outcomes of different trials are independent.

---

9. What is the Poisson distribution and where is it applied?

Ans: The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space, assuming these events occur with a known constant mean rate and independently of the time since the last event.

---

10. What is a continuous uniform distribution?

Ans: A continuous uniform distribution is a probability distribution in which all intervals of the same length on the distribution's support are equally probable. It is characterized by two parameters, aa and bb, which are the minimum and maximum values, respectively.

---

11. What are the characteristics of a normal distribution?

Ans: A normal distribution is symmetric, bell-shaped, and fully characterized by its mean and standard deviation. It describes many natural phenomena and measurement errors, where most observations cluster around the mean and probabilities for values further from the mean drop off symmetrically in both directions.

---


12. What is the standard normal distribution, and why is it important?

Ans: The standard normal distribution is a special case of the normal distribution with a mean of zero and a standard deviation of one. It is important because any normal distribution can be transformed into a standard normal distribution, and it is widely used in statistical methods, especially hypothesis testing.

---

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Ans: The Central Limit Theorem states that the sampling distribution of the sample means approaches a normal distribution as the sample size gets larger, regardless of the shape of the population distribution. This is critical in statistics because it justifies the assumption of normality in many statistical tests.

---

14. How does the Central Limit Theorem relate to the normal distribution?

Ans: The Central Limit Theorem explains why many results in statistics are normally distributed, as it states that the distribution of the sample mean approximates a normal distribution as the sample size increases, regardless of the population's distribution.

15. What is the application of Z statistics in hypothesis testing?

Ans: Z statistics are used in hypothesis testing to determine whether to reject the null hypothesis, comparing the Z statistic to standard normal distribution values to assess the likelihood of observing a sample statistic if the null hypothesis is true.

---

16. How do you calculate a Z-score, and what does it represent?

Ans: A Z-score is calculated as Z=(X−μ)σZ=σ(X−μ)​, where XX is the value, μμ is the mean, and σσ is the standard deviation. It represents how many standard deviations an element is from the mean.

---

17. What are point estimates and interval estimates in statistics?

Ans: Point estimates provide a single value estimate of a parameter, while interval estimates provide a range of values, expressed as an interval, that is likely to contain the parameter.

---

18. What is the significance of confidence intervals in statistical analysis?

Ans: Confidence intervals provide a range of values which likely contain the population parameter. The level of confidence expresses the degree of certainty associated with the interval.

---

19. What is the relationship between a Z-score and a confidence interval?

Ans: A Z-score helps calculate the boundaries of a confidence interval in a normal distribution, determining how far the interval extends from the mean, based on the desired confidence level.

---

20. How are Z-scores used to compare different distributions?
Ans: Z-scores standardize scores from different distributions, allowing for comparison by transforming them to a common scale with a mean of zero and a standard deviation of one.
---

21. What are the assumptions for applying the Central Limit Theorem?

Ans: The main assumptions are that the samples are independent, and the sample size is sufficiently large, typically n≥30n≥30, although the theorem applies regardless of the distribution's shape, given these conditions.

---

22. What is the concept of expected value in a probability distribution?

Ans: The expected value is the long-run average value of repetitions of the experiment it represents. It is calculated as the weighted average of all possible values.

---

23. How does a probability distribution relate to the expected outcome of a random variable?

Ans: A probability distribution provides a complete description of the probability of all possible outcomes of a random variable, and the expected value is a summary measure of the center of the distribution, providing a sense of the 'average' outcome over many trials.

---

2)Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

```python
import matplotlib.pyplot as plt


values = np.arange(1, 7)
pmf = [1/len(values)] * len(values)

plt.bar(values, pmf)
plt.title('Discrete Uniform Distribution PMF')
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.show()

```

3)Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
def bernoulli_pdf(x, p):
    return p if x == 1 else 1 - p

print("PDF when x=1:", bernoulli_pdf(1, 0.5))
print("PDF when x=0:", bernoulli_pdf(0, 0.5))
```

9)Write a Python function to calculate Z-scores from a dataset and plot them?

```python
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

data = np.random.normal(loc=50, scale=10, size=1000)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, color='orange', edgecolor='black')
plt.title('Z-scores Histogram')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.show()
```

15. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

20. Create a confidence interval for a dataset using Python and interpret the result.

```python
import scipy.stats as stats

data = np.random.normal(loc=50, scale=10, size=100)
mean = np.mean(data)
std = np.std(data, ddof=1)
confidence_interval = stats.norm.interval(0.95, loc=mean, scale=std/np.sqrt(len(data)))

print("95% Confidence Interval:", confidence_interval)
```

21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

data = np.random.normal(loc=50, scale=10, size=1000)

mean = np.mean(data)
std = np.std(data, ddof=1) / np.sqrt(len(data))


confidence_interval = stats.norm.interval(0.95, loc=mean, scale=std)

print("Mean of the data:", mean)
print("95% Confidence Interval:", confidence_interval)


print("\nInterpretation:")
print(f"The true mean of the population is likely between {confidence_interval[0]:.2f} and {confidence_interval[1]:.2f} with 95% confidence.")
```

22. Calculate and visualize the probability density function (PDF) of a normal distribution.

```python

mean = 0
std = 1

x = np.linspace(mean - 3*std, mean + 3*std, 100)
y = stats.norm.pdf(x, mean, std)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Standard Normal PDF')
plt.title('Probability Density Function of a Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

23. Calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

```python
from scipy.stats import poisson

mu = 3
x = np.arange(0, 10)
cdf = poisson.cdf(x, mu)

plt.step(x, cdf, where='post')
plt.title('Poisson Distribution CDF')
plt.xlabel('Number of Events')
plt.ylabel('Cumulative Probability')
plt.show()
```

26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
data = np.random.normal(loc=50, scale=10, size=1000)
data_range = np.ptp(data)  # Peak to peak (max - min)
iqr = np.percentile(data, 75) - np.percentile(data, 25)

print("Range:", data_range)
print("Interquartile Range (IQR):", iqr)
```
