### Q.1) What is the difference between a function and a method in Python?

Function: A standalone block of code designed to perform a specific task, defined using the def keyword.

Method: A function associated with an object, defined within a class, and intended to operate on instances of that class.

```python
#Example of Function:
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Outputs: Hello, Alice!
```

```python
#Example of Method

class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

g = Greeter()
print(g.greet("Bob"))  # Outputs: Hello, Bob!
```

### Q.2)Explain the concept of function arguments and parameters in Python?

Parameters are the names used in a function definition that specify what kind of data a function can accept.
Arguments are the actual data you provide to a function when you call it. They are mapped to the corresponding parameters in the function.

```python
def add(a, b):  # 'a' and 'b' are parameters
    return a + b

result = add(5, 3)  # 5 and 3 are arguments
print(result)
```

### Q.3)What are the different ways to define and call a function in Python?

```python
# Standard Function Definition
#Define functions using the def keyword followed by the function name and parameters.
#Call the function by using its name followed by parentheses containing any necessary arguments.

#Example:

def add(a, b):
    return a + b
result = add(5, 3)
print(result)
```

```python
# Functions with Default Parameters
#Define functions with default values for parameters,
#which are used if no corresponding arguments are passed during the function call.

#Example:

def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Alice")
greet("Bob", "Howdy")
```

```python
#Arbitrary Arguments

#Use *args for arbitrary positional arguments and **kwargs for arbitrary keyword arguments
#when the number of arguments is not known beforehand.

#Example using *args:

def print_names(*names):
    for name in names:
        print(name)

print_names("Alice", "Bob", "Charlie")
```

```python
#Lambda Functions

#Define small anonymous functions using the lambda keyword.
#Lambda functions can have any number of arguments but only one expression.

#Example:

square = lambda x: x * x
print(square(4))
```

```python
#Generator Functions

#Generator functions use the yield keyword to return data one piece at a time,
# pausing after each yield and resuming with the next call.

#Example:

def count_down(start):
    while start > 0:
        yield start
        start -= 1

for number in count_down(5):
    print(number)
```

### Q.6)Explain the concept of generators in Python and how they are defined?

### Q.7)What are the advantages of using generators over regular functions?

Advantages of Generators:

    * Memory Efficiency: They generate items on-the-fly and use very little memory.
    * Lazy Execution: They produce items only when necessary, optimizing resource usage.
    * State Preservation: They maintain their state between iterations automatically.
    * Simplify Code: They allow for cleaner, more readable code, especially in looping constructs.
    * Pipeline Creation: They can be chained together for processing streams of data efficiently.
    * Infinite Sequences: They are capable of producing infinite data sequences without filling up memory.




```python
def read_large_file(file_name):
    with open(file_name, 'r') as file:
        for line in file:
            yield line.strip()

# Usage of the generator to process lines as they are read
for line in read_large_file("largefile.txt"):
    print(line)  # Process each line one at a time without loading the entire file into memory
```

### Q.9) Explain the purpose and usage of the `map()` function in Python?

The map() function in Python is used to apply a function to each item in an iterable (like a list or tuple), returning a map object (an iterator) with the results. It's a way to transform data efficiently and succinctly without using explicit loops.

```python
#Usage :
# The map() function takes at least two parameters:
# A function - This can be a standard function, a lambda function, or any callable object.
# An iterable (like a list, tuple, etc.) - map() applies the function to each element of this iterable.

#syntax

map(function, iterable)
#You can also pass multiple iterables to map(), in which case the function must take as many arguments as there are iterables. Hereâ€™s how it works:
#Example with Single Iterable

# Doubling each number in a list using a lambda function
numbers = [1, 2, 3, 4, 5]
doubled_numbers = map(lambda x: x * 2, numbers)
print(list(doubled_numbers))  # Output: [2, 4, 6, 8, 10]

# Example with Multiple Iterables

# Adding elements from two lists
list1 = [1, 2, 3]
list2 = [4, 5, 6]
result = map(lambda x, y: x + y, list1, list2)
print(list(result))  # Output: [5, 7, 9]
```

### Q.10) What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

* map()

    Applies a function to every item of an iterable, returning an iterator with the transformed items.

* filter()

    Applies a function that returns either True or False to each item of an iterable, returning an iterator with items that returned True.

* reduce()

    Applies a binary function cumulatively to the items of an iterable, reducing them to a single value. Unlike map() and filter(), reduce() is not a built-in function and must be imported from the functools module.

```python
#Example Map
numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
```

```python
#Example Filter
numbers = [1, 2, 3, 4]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
```

```python
#Example Reduce
from functools import reduce
numbers = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, numbers)
```

### Q.1) Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in?

```python
def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)
```

### Q.4)Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(number):
    if number < 2: return False
    for i in range(2, number):
        if number % i == 0: return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
```

### Q.5)Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

# Example usage:
for num in fibonacci(10):
    print(num)
```


### Q.6) Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for power in powers_of_2(5):
    print(power)
```
